```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import GridSearchCV
import seaborn as sns
from xgboost import plot_importance, plot_tree

import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable

from skforecast.ForecasterAutoreg import ForecasterAutoreg
from skforecast.ForecasterAutoregMultiOutput import ForecasterAutoregMultiOutput
from skforecast.model_selection import grid_search_forecaster
from skforecast.model_selection import backtesting_forecaster

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
read_file = pd.read_excel ("cnn.xlsx")
  
# Write the dataframe object
# into csv file
read_file.to_csv ("cnn.csv", 
                  index = None,
                  header=True)
    
df = pd.read_csv('cnn.csv', header = None)
df
```

```python
print(f'Number of rows with missing values: {df.isnull().any(axis=1).mean()}')
```

```python
df.columns = df.iloc[0] 
df
```

```python
df = df[1:]
df
```

```python
end_train = '410'
end_validation = '520'
data_train = df.loc[: end_train, :]
data_val   = df.loc[end_train:end_validation, :]
data_test  = df.loc[end_validation:, :]
print(f"Dates train      : {data_train.index.min()} --- {data_train.index.max()}  (n={len(data_train)})")
print(f"Dates validacion : {data_val.index.min()} --- {data_val.index.max()}  (n={len(data_val)})")
print(f"Dates test       : {data_test.index.min()} --- {data_test.index.max()}  (n={len(data_test)})")
```

```python
forecaster = ForecasterAutoreg(
                regressor = XGBRegressor(random_state=123),
                lags = 12
                )

forecaster
```

```python
param_grid = {
    'n_estimators': [100, 500],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1]
    }

# Lags used as predictors
lags_grid = [10, 20]

for level in range(1,50):
    results_grid = grid_search_forecaster(
            forecaster         = forecaster,
            y                  = df.loc[:end_validation, level], # Train and validation data
            param_grid         = param_grid,
            lags_grid          = lags_grid,
            steps              = 25,
            refit              = False,
            metric             = 'mean_squared_error',
            initial_train_size = int(len(data_train)), # Model is trained with trainign data
            fixed_train_size   = False,
            return_best        = True,
            verbose            = False
            )
```

```python
results_grid.head(5)
```

```python
for level in range(1,50):
    metric, predictions = backtesting_forecaster(
        forecaster = forecaster,
        y          = df[level],
        initial_train_size = len(df.loc[:end_validation]),
        fixed_train_size   = False,
        steps      = 25,
        refit      = False,
        metric     = 'mean_squared_error',
        verbose    = False # Change to True to see detailed information
        )

print(f"Backtest error: {metric}")
```
