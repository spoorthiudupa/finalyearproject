# Hair removal process

### This file has the hair removal process for a sample image. The steps performed are:

1. Using this BSD, hair removal techniques are performed to ease the process of extracting features from samples. 
2. The image is first converted to a grayscale image and then it is passed to a Blackhat filter which enhances dark regions of interest. 
3. This is then passed to a thresholding technique. 
4. Finally, the inpainting algorithm is applied to this image. 
5. The regions with hair are removed from the original image and it outputs an image with no hair. 

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Plotting a benign mole with hair

```python
img = mpimg.imread("final/benign/1 (56).jpg")
imgplot = plt.imshow(img)
plt.show()
```

This perfoms all the steps to remove hair. The intermediate results are stored in hair_removal folder. 

```python
src = cv2.imread("final/benign/1 (56).jpg")

grayScale = cv2.cvtColor( src, cv2.COLOR_RGB2GRAY )
cv2.imwrite('hair_removal/grayScale_sample1.jpg', grayScale, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

kernel = cv2.getStructuringElement(1,(17,17))

blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)
cv2.imwrite('hair_removal/blackhat_sample1.jpg', blackhat, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

ret,thresh2 = cv2.threshold(blackhat,10,255,cv2.THRESH_BINARY)
print( thresh2.shape )
cv2.imwrite('hair_removal/thresholded_sample1.jpg', thresh2, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

dst = cv2.inpaint(src,thresh2,1,cv2.INPAINT_TELEA)
cv2.imwrite('hair_removal/InPainted_sample1.jpg', dst, [int(cv2.IMWRITE_JPEG_QUALITY), 90])
```

The final image with  hair is printed

```python
img = mpimg.imread("hair_removal/InPainted_sample1.jpg")
imgplot = plt.imshow(img)
plt.show()
```
